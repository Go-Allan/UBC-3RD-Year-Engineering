```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

i = 0
inc = 375 # number of increments 
n = 0.15 # increment size
x = np.zeros(inc)
startval = 4.01

while i < inc:
    x[0] = startval
    x[i] = x[i-1] + n
    i += 1
    
#print(x)

def func(wl_NaOH_new):
    # Variable Definition
    wf_NaOH = 0.0301 #weight fraction of NaOH in feed
    wl_NaOH = wl_NaOH_new #weight fraction of NaOH in liquor (0.5 in original question)
    mf = 9580 #[kg/hr] mass rate of feed
    Tf = 27 #[C] temperature of feed
    P_S = 821 #[kPa] Pressure of Steam
    T_S = 171.5 #[C] Temperature of Steam 
    PTank = 6.89 #[KPa]
    u = 1547 #[W/Km^2]
    MW_H2O = 18.0 #[g/mol]
    MW_NaOH = 40 #[g/mol]
    Cp = 4184 #[J/kg]

    # Saturation Pressure and Temperature Data
    P = [5.0,6.0,7.0,8.0,9.0,10.0,12.0,14.0,16.0,18.0,20.0] #[KPa]
    T = [32.90,36.18,39.03,41.54,43.79,45.83,49.45,52.58,55.34,57.83,60.09] #[C]

    #find mole fraction of solute (xNaOH)
    wl_H2O = 1-wl_NaOH #weight fraction of water in liquor
    xl_NaOH = (wl_NaOH/MW_NaOH)/((wl_NaOH/MW_NaOH)+(wl_H2O/MW_H2O))

    #find satruation pressure of water
    Psat = PTank/(1-xl_NaOH) #[Kpa] rearranged from PTank = (1-xsolute)*Psat
    #print("Saturation Pressure is:",Psat)

    #Find Saturation Temperature from Psat 
    Ttank = np.interp(Psat,P,T) #[C]
    #print("Temp in tank is:",Ttank)

    # find the mass of liquor
    ml = (mf * wf_NaOH)/wl_NaOH  #rearranged from: mf*wf_NaOH = ml*wl_NaOH
    #print("mass of liquor is:",ml)

    #find mass of vapour
    mv = mf - ml 
    #print("mass of vapour is:",mv)

    #Find enthalpy of vaporization in tank (H2O)
    Hv_vap = np.interp(Psat, [9,10],[2397.9,2392.9]) #[kJ/kg]
    #print("Hvap is:", Hv_vap)

    #h1-h2
    H1_2 = np.interp(T_S,[170.41,175.35],[2047.44,2030.47]) #[kJ/kG]
    #print("h1-h2 is:",H1_2)

    #solve for hv - hf
    HV_F = Hv_vap + (Cp*(Ttank - Tf))/1000 #[kJ/kg] # assume vapor is superheated
    #print("hv-hf is:",HV_F)

    #solve for h_liq - hf 
    delta_hqhf = Cp*(Ttank - Tf)/1000 #[kJ/kg]
    #print("hl-hf is:",delta_hqhf)

    #solve for mass of steam (m1)
    m1 = (mv*HV_F+ml*delta_hqhf)/H1_2
    #print("Mass flow is:",m1)

    #economy
    econ = mv/m1
    #print("economy is:",econ)
    
    return m1
    
y = np.zeros(inc)    

for m in x:
    y = func(x)
    
#print(y)

plt.plot(x,y,'b-')
plt.xlabel('Concentration of NaOH\nin Liquor Stream (wt%)', fontsize = 14)
plt.ylabel('Mass Flow Rate of \nSteam Required (kg/h)', fontsize = 14)
plt.title('Flow Rate of Steam Required vs Concentration \nof NaOH in a Single Effect Evaporator', fontsize = 18)
plt.tight_layout()
plt.show()
```
