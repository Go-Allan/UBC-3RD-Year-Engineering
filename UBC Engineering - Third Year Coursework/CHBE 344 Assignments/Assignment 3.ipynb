```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def funky(x,t):
    x1,x2 = x # where x1 = x(t), x2 = d(x1)/dt
    dxdt = [x2, -3] # (derivative of x2) = (d^2)x/dt^2
    return dxdt

x0 = [5, 15]

t = np.linspace(0, 10, 40)
sol = odeint(funky, x0, t)

analytical1 = -3/2*t**2+15*t+5
analytical2 = -3*t+15

# Show X vs t graph
plt.figure(figsize=(4.5,4))
plt.plot(t, sol[:, 0], 'ks', label='Numerical',ms = 8)
plt.plot(t, analytical1, 'r', label='Analytical',lw = 3)
plt.xlabel("time", fontsize = 22)
plt.xticks(np.arange(0,11,5))
plt.ylabel('X', fontsize = 22)
plt.yticks(np.arange(0,51,10))
plt.legend()
plt.show()

# Show dx/dt graph
plt.figure(figsize=(4.5,4))
plt.plot(t, sol[:, 1], 'ks', label='Numerical',ms = 8)
plt.plot(t, analytical2, 'r', label='Analytical', lw = 3)
plt.xlabel("time", fontsize = 22)
plt.xticks(np.arange(0,11,5))
plt.ylabel('dx/dt', fontsize = 22)
plt.yticks(np.arange(-15,16,5))
plt.legend()
plt.show()
```

```python
import math as m

E0 = 14700 # [J/kg]
Li0 = 0.0625 # [m]
Lf0 = 0.0105 # [m]

# Calculating C Values for Each of Rittinger, Kick, and Bond's Laws
c_rit = E0/(1/Lf0-1/Li0)
c_kick = E0/m.log(Li0/Lf0) # m.log is natural logarithm
c_bond = E0/2/(1/m.sqrt(Lf0)-1/m.sqrt(Li0))
#print(c_rit,c_kick,c_bond)

Li = 0.075 # [m]
Lf = np.arange(0.050,0.004,-0.001)
#print(Lf)

# i)
E_rit = []

i = 0
while i < len(Lf): #for loop error?...
    E_rit.append(c_rit*(1/Lf[i]-1/Li))
    i += 1

# ii)
E_kick = []

i = 0
while i < len(Lf):
    E_kick.append(c_kick*m.log(Li/Lf[i]))
    i += 1
    
# iii)
E_bond = []

i = 0
while i < len(Lf):
    E_bond.append(c_bond*2*(1/m.sqrt(Lf[i])-1/m.sqrt(Li)))
    i += 1

# Plotting
plt.plot(Lf, E_rit,'r-', label = 'Rittinger')
plt.plot(Lf,E_kick,'b.', label = 'Kick')
plt.plot(Lf,E_bond,'g--', label = 'Bond')
plt.ylabel("Energy Required [J/kg]", fontsize=18)
plt.xlabel("Final Particle Size [m]", fontsize=18)
plt.title("Energy Required to Achieve Final Particle Size\nBased on Three Models"
         ,fontsize = 20)
# plt.gca().invert_xaxis() # Decreasing particle size increases energy required*
plt.legend()
plt.show()

#print(E_rit)
#print(E_kick)
#print(E_bond)
```

```python
import csv
from scipy.interpolate import interp1d
import numpy as np

# READ FILE
aperture = []
feed = []
accepts = []
rejects = []

#make sure to have file for reading in same folder as this code (else have to call file location)
with open('HW3-2 Data.csv','r') as file: 
    data = csv.reader(file, delimiter = ',')
    file.readline() #skip first two lines with no useful data
    file.readline()
    for row in data:
        aperture.append(float(row[0]))
        feed.append(float(row[1]))
        accepts.append(float(row[2]))
        rejects.append(float(row[3]))

# INTERPOLATE SIEVE SIZE
sieve_size = (1-12*0.0401)/12*25400 # computes length of opening in sieve based on question [microns]
print("sieve size =",sieve_size,'microns')

for i in range(0,len(aperture)):
    if aperture[i] < sieve_size and aperture[i+1] > sieve_size:
        print(aperture[i], aperture.index(aperture[i]))

# Sum all values above the two sieves on either side of the industrial sieve
xfL = 0
xfS = 0
xaL = 0
xaS = 0
xrL = 0
xrS = 0
appL = 0
appS = 0

for i in range(0, len(aperture)):
    if aperture[i] > sieve_size:
        
        xfL += feed[i]
        xaL += accepts[i]
        xrL += rejects[i]
        appL += aperture[i]
        
    if aperture[i] >= (aperture[18]):
        xfS += feed[i] 
        xaS += accepts[i]
        xrS += rejects[i]
        appS += aperture[i]

# Set up interpolation
app = [aperture[18],aperture[19]]
f = [xfS,xfL]
acc = [xaS,xaL]
rej = [xrS,xrL]

# interpolate the mass fractions sitting on top of our industrial sieve
Xf = np.interp(sieve_size,app,f) 
Xa = np.interp(sieve_size,app,acc)
Xr = np.interp(sieve_size,app,rej)

# Print values to sanity check
print(xfS,xaS,xrS)
print(Xf,Xa,Xr)
print(xfL,xaL,xrL)
      
# mass fraction of small particles is 1 - mass fraction of larger particles
# (the mass fraction sitting on top of the industrial sieve)
Xfs = 1 - (Xf)
Xas = 1 - (Xa)
Xrs = 1 - (Xr)

eff = ((Xas)*(Xfs-Xrs)*(1-Xrs)*(Xas-Xfs))/((Xfs)*(1-Xfs)*(Xas-Xrs)**2)
print("effectiveness =",eff)
```
